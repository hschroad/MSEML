```python
# Load in the data
raw_df = np.loadtxt(open("df_no_duplicates_clean_2.0.csv", "r"), delimiter = ',', dtype = str, skiprows = 0)

# Get the labels
df_labels = raw_df[0,1:]
internal_ids = raw_df[1:,1]

# Make the dataframe
df = pd.DataFrame(raw_df[1:,1:], columns = df_labels, index = internal_ids)

# Convert to numeric values
df = df.iloc[:,1:].copy().apply(pd.to_numeric)

data = df.iloc[:,1:].copy().values

df.head()
```
